The six variables that improved were artificially collapsed in the original

across all observations. The cleaned dataset now reflects the true variability
of real particles, making these variables genuinely informative for clustering.

## Extractor corrections and dataset regeneration

During the exploratory analysis of `particle_features_3.csv`, two bugs were
identified in the feature extractor (`analysis.py`) that were producing
invalid values for a subset of particles:

### Bug 1 — `convexity_defects` with negative values

The original formula computed:

```python
convexity_defects = hull_area - area
```

where `hull_area` was calculated using `cv2.contourArea(hull)` on global
pixel coordinates, while `area` came from `region.area` (skimage). For very
small particles, these two methods produce inconsistent results due to
coordinate rounding, causing `hull_area < area` and therefore
`convexity_defects < 0`, which is geometrically impossible — the convex
hull area must always be ≥ the particle area.

In `particle_features_3.csv` this affected the majority of particles:
`convexity_defects` had a median of -3.0 and a minimum of -49.

### Bug 2 — `fractal_dimension` returning `-inf`

The original formula:

```python
fractal_dimension = log(perimeter) / log(area)  if area > 1  else 0
```

For particles where `perimeter = 0` (skimage cannot compute the contour
for certain degenerate geometries even with `area ≥ 3`), this produced
`log(0) = -inf`. There were 143 such cases in `particle_features_3.csv`.

### Fix applied to `analysis.py`

Both bugs were corrected in the extractor. `convexity_defects` is now
computed from the particle's local contour (`region.image`) rather than
global coordinates, ensuring it is always ≥ 0. `fractal_dimension` now
returns `NaN` instead of `-inf` for undefined cases, which is the
statistically correct representation of a missing value.

### Impact on the dataset

The corrected extractor was re-run on all images with `min_area=3`,
producing `particle_features_3.1.csv`. The dataset was reduced from
222,838 to 197,536 particles (-25,302, -11.4%), as the corrected extractor
identified additional degenerate particles that the original had retained
with invalid metric values. `fractal_dimension` now has 121 NaN values
(down from 143 `-inf`) and `convexity_defects` is strictly ≥ 0 across
all particles.

This version of the dataset (`particle_features_3.1.csv`) is used for
all subsequent analysis.
