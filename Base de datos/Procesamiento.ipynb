```python
import os
from PIL import Image, ExifTags
try:
    from pillow_heif import register_heif_opener
    register_heif_opener()
except ImportError:
    print("Nota: Si usas fotos de iPhone (.HEIC), instala: pip install pillow-heif")

def analizar_particulas_movil(directorio):
    extensiones = ('.jpg', '.jpeg', '.png', '.heic', '.heif', '.tif')
    
    if not os.path.exists(directorio):
        print(f"❌ Error: La ruta no existe: {directorio}")
        return

    print(f"{'ARCHIVO':<45} | {'ANCHO':<8} | {'ALTO':<8} | {'MP'}")
    print("-" * 75)

    for nombre in os.listdir(directorio):
        if nombre.lower().endswith(extensiones):
            ruta_full = os.path.join(directorio, nombre)
            try:
                with Image.open(ruta_full) as img:
                    # Corregir orientación según metadatos EXIF del móvil
                    try:
                        for orientation in ExifTags.TAGS.keys():
                            if ExifTags.TAGS[orientation] == 'Orientation':
                                break
                        exif = dict(img._getexif().items())
                        if exif[orientation] == 3: img = img.rotate(180, expand=True)
                        elif exif[orientation] == 6: img = img.rotate(270, expand=True)
                        elif exif[orientation] == 8: img = img.rotate(90, expand=True)
                    except (AttributeError, KeyError, IndexError):
                        pass # No tiene EXIF o ya está correcta

                    w, h = img.size
                    mp = (w * h) / 1_000_000
                    print(f"{nombre[:43]:<45} | {w:<8} | {h:<8} | {mp:.2f}")
            except Exception as e:
                print(f"Error procesando {nombre}: {e}")

# --- RUTA DE TU PROYECTO ---
ruta_pfg = r"C:\Users\Usuario\Documents\GitHub\PFG\Analizador\pollution-particle-analyzer-main\a"

analizar_particulas_movil(ruta_pfg)
```

```python
# -*- coding: utf-8 -*-
"""
outlier_inspection.py
======================
Visualiza las partículas con valores extremos para decidir
si son ruido de segmentación o partículas reales.

Pega esto en tu notebook ANTES del preprocesamiento.
Necesitas tener acceso a las imágenes originales.
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from PIL import Image
import os

# ─────────────────────────────────────────────────────────────────
# CONFIGURACIÓN — ajusta estas rutas
# ─────────────────────────────────────────────────────────────────

RUTA_IMAGENES = r"C:\Users\Usuario\Documents\GitHub\PFG\Analizador\pollution-particle-analyzer-main\a"
CSV_PATH      = r"C:\Users\Usuario\Documents\GitHub\PFG\Analizador\pollution-particle-analyzer-main\output\particle_features.csv"

df = pd.read_csv(CSV_PATH)
df['fractal_dimension'].replace(-np.inf, 0, inplace=True)

# ─────────────────────────────────────────────────────────────────
# FUNCIÓN PRINCIPAL: mostrar partículas en los extremos de una variable
# ─────────────────────────────────────────────────────────────────

def inspect_extremes(df, variable, n=8, percentile_high=99, percentile_low=1,
                     ruta_imagenes=RUTA_IMAGENES, crop_margin=20):
    """
    Muestra las n partículas con valores más altos y más bajos de `variable`.
    Recorta la región de la partícula de la imagen original usando bbox.

    Requiere columnas en df: image_name, bbox_x, bbox_y, bbox_w, bbox_h
    (ajusta los nombres si los tuyos son distintos)
    """

    # Detectar columnas de bounding box disponibles
    bbox_cols = [c for c in df.columns if "bbox" in c.lower() or "bounding" in c.lower()]
    print(f"Columnas bbox encontradas: {bbox_cols}")

    val_high = df[variable].quantile(percentile_high / 100)
    val_low  = df[variable].quantile(percentile_low  / 100)

    df_high = df[df[variable] >= val_high].nlargest(n, variable)
    df_low  = df[df[variable] <= val_low ].nsmallest(n, variable)

    print(f"\n{variable}:")
    print(f"  p{percentile_low}  = {val_low:.4f}   →  {len(df_low)} partículas por debajo")
    print(f"  p{percentile_high} = {val_high:.4f}   →  {len(df_high)} partículas por encima")

    for label, subset in [("VALORES ALTOS", df_high), ("VALORES BAJOS", df_low)]:
        fig, axes = plt.subplots(2, n, figsize=(n * 2.5, 5))
        fig.suptitle(f"{variable} — {label}", fontsize=12, fontweight="bold")

        for i, (_, row) in enumerate(subset.iterrows()):
            val = row[variable]
            img_name = row.get("image_name", "")
            img_path = os.path.join(ruta_imagenes, img_name)

            # Fila superior: crop de la partícula (si existe la imagen)
            ax_img = axes[0, i]
            if os.path.exists(img_path) and bbox_cols:
                try:
                    img = Image.open(img_path).convert("L")
                    # Intenta leer bbox — adapta los nombres de columna si hace falta
                    x = int(row.get("bbox_x", row.get("bbox_minr", 0)))
                    y = int(row.get("bbox_y", row.get("bbox_minc", 0)))
                    w = int(row.get("bbox_w", row.get("bbox_maxr", 50) - x))
                    h = int(row.get("bbox_h", row.get("bbox_maxc", 50) - y))

                    # Crop con margen
                    x1 = max(0, x - crop_margin)
                    y1 = max(0, y - crop_margin)
                    x2 = min(img.width,  x + w + crop_margin)
                    y2 = min(img.height, y + h + crop_margin)
                    crop = img.crop((x1, y1, x2, y2))
                    ax_img.imshow(crop, cmap="gray")
                except Exception as e:
                    ax_img.text(0.5, 0.5, f"Error\n{e}", ha="center", va="center",
                                fontsize=7, transform=ax_img.transAxes)
            else:
                ax_img.text(0.5, 0.5, "Sin imagen", ha="center", va="center",
                            fontsize=8, color="gray", transform=ax_img.transAxes)

            ax_img.set_title(f"{val:.2f}", fontsize=8)
            ax_img.axis("off")

            # Fila inferior: valores de otras variables clave para contexto
            ax_info = axes[1, i]
            info = (
                f"diam: {row.get('equivalent_diameter', '?'):.1f}\n"
                f"round: {row.get('roundness', '?'):.2f}\n"
                f"solid: {row.get('solidity', '?'):.2f}\n"
                f"holes: {row.get('holes', '?')}\n"
                f"euler: {row.get('euler_number', '?')}"
            )
            ax_info.text(0.5, 0.5, info, ha="center", va="center",
                         fontsize=7, family="monospace", transform=ax_info.transAxes)
            ax_info.axis("off")

        plt.tight_layout()
        plt.show()


# ─────────────────────────────────────────────────────────────────
# INSPECCIÓN DE VARIABLES PROBLEMÁTICAS
# Ejecuta celda a celda para ir decidiendo
# ─────────────────────────────────────────────────────────────────

# 1. Roundness — valores >1 son imposibles matemáticamente
print("="*60)
print("ROUNDNESS — máximo teórico = 1")
print(f"Valores > 1: {(df['roundness'] > 1).sum()} filas")
print(f"Valores > 5: {(df['roundness'] > 5).sum()} filas")
print(df['roundness'].describe())
inspect_extremes(df, "roundness", n=8, percentile_high=99)

# ─────────────────────────────────────────────────────────────────

# 2. Holes y euler_number — ¿ruido o partículas reales con agujeros?
print("\n" + "="*60)
print("HOLES — distribución de valores altos")
print(df['holes'].value_counts().sort_index().tail(15))
inspect_extremes(df, "holes", n=8, percentile_high=95)

print("\n" + "="*60)
print("EULER NUMBER")
print(df['euler_number'].describe())
inspect_extremes(df, "euler_number", n=8, percentile_low=5)

# ─────────────────────────────────────────────────────────────────

# 3. Equivalent diameter — ¿cuánto es un pixel en tu imagen?
print("\n" + "="*60)
print("EQUIVALENT DIAMETER")
print(df['equivalent_diameter'].describe())
print(f"\nDistribución por rangos:")
bins = [0, 2, 5, 10, 20, 50, 100, 1000]
print(pd.cut(df['equivalent_diameter'], bins=bins).value_counts().sort_index())
inspect_extremes(df, "equivalent_diameter", n=8, percentile_high=99)

# ─────────────────────────────────────────────────────────────────

# 4. Convexity defects — valores extremos
print("\n" + "="*60)
print("CONVEXITY DEFECTS")
print(df['convexity_defects'].describe())
inspect_extremes(df, "convexity_defects", n=8, percentile_high=99)

# ─────────────────────────────────────────────────────────────────

# 5. RESUMEN: cuántos outliers hay por variable con IQR×1.5 y IQR×3
print("\n" + "="*60)
print("RESUMEN DE OUTLIERS POR VARIABLE (IQR)")
print(f"{'Variable':<25} {'IQR×1.5':>10} {'IQR×3':>10} {'% IQR×1.5':>12}")
print("-" * 60)

numeric_cols = df[df.columns.difference(["image_name"])].select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
    IQR = Q3 - Q1
    if IQR == 0:
        continue
    n_15 = ((df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)).sum()
    n_3  = ((df[col] < Q1 - 3.0*IQR) | (df[col] > Q3 + 3.0*IQR)).sum()
    pct  = n_15 / len(df) * 100
    print(f"{col:<25} {n_15:>10} {n_3:>10} {pct:>11.1f}%")
```
