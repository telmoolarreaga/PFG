## Why min_area=2 is still insufficient

After raising `min_area` from 0 to 2 in `config.py`, the dataset was reduced
from 1,702,864 to 332,270 particles (-80.5%). However, inspection reveals that
**123,324 particles (37.1%) still have `area = 2px²` with `perimeter = 0`**.

These particles share the same root cause as the 1px² particles eliminated
in the previous step: skimage cannot compute a valid contour for particles of
only 2 connected pixels, resulting in `perimeter = 0`. This cascades into
multiple invalid features — all are exactly 0 or undefined across every
geometric descriptor:

- `roundness = 0` — formula `4π·area/perimeter²` → division by zero
- `compactness = 0` — formula `perimeter²/area` → numerator is 0
- `fractal_dimension = 0` — formula `log(perimeter)/log(area)` → `log(0)`,
  numerically collapsed to 0 in this extractor version
- `convexity_defects = -2` — constant for all 2px² particles,
  result of OpenCV hull calculation failing on degenerate contours

Additionally, `convexity_defects < 0` affects 321,773 particles (96.8%
of the dataset), indicating a broader bug in the original extractor that
computes hull area incorrectly for very small particles. This was corrected
in extractor version 3.1.

**Decision: raise `min_area` from 2 to 3** in `config.py` and regenerate
the CSV. This eliminates the 123,324 degenerate 2px² particles while
preserving all particles of 3px² and above, which have valid geometric
metrics. The resulting dataset will contain approximately 208,946 particles
(-37.1% from current).

## Note on dataset size reduction

The initial estimate predicted eliminating ~53% of particles by raising
`min_area` to 2, based on the proportion of 1px² particles observed in
the original CSV. The actual reduction was **80.5%** (from 1,702,864 to
332,270 particles).

The larger-than-expected reduction occurred because the extractor applies
all filters in `config.py` simultaneously during regeneration — not just
`min_area`. Particles that previously passed the area filter but failed
other criteria (`min_solidity=0.3`, `max_aspect_ratio=4.0`, `max_feret=50`)
were also excluded. In the original CSV, most degenerate small particles
had default metrics of 0 or 1, which allowed them to pass all filters
despite being noise. With `min_area=2`, the stricter size threshold
exposed these additional filtering effects.
