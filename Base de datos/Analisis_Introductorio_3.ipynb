After setting min_area=3 in config.py, 273 particles remain with
perimeter = 0 (areas of 4 and 5px²), where skimage failed to compute
the contour correctly.

These particles are removed because:
- perimeter = 0 is geometrically impossible for any real particle
- All derived metrics that depend on perimeter are invalid:
  roundness = 0, compactness = 0, fractal_dimension = -inf

These are eliminated directly in the notebook as they are too few (0.1%)
to justify raising min_area further and losing valid particles.

### Variables correlation

```python
# Numerical Columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

#  Correlation matrix
corr_matrix = numeric_df.corr()

#  Highly correlated pairs
threshold = 0.9
high_corr = (corr_matrix.abs() > threshold) & (corr_matrix.abs() < 1.0)  
print("Variables highly correlated (> 0.9):")
for col in high_corr.columns:
    correlated_cols = high_corr.index[high_corr[col]].tolist()
    if correlated_cols:
        print(f"{col} ↔ {correlated_cols}")

# Correlation heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```



After removing degenerate particles (`min_area=3`, `perimeter=0` filter),
the correlation matrix shows only minor changes:

- The size block (`area`, `perimeter`, `major_axis`, `equivalent_diameter`...)
  shows slightly stronger correlations, as the geometric relationships between
  variables emerge more clearly without degenerate particles.
- `fractal_dimension` shows a more pronounced negative correlation with
  `convexity_defects` and `hull_ratio`, previously masked by the `-inf` values.

However, the fundamental structure remains unchanged — the relationships
between the variables selected for clustering are consistent across both
datasets. Therefore, **the original variable selection is maintained**:

- **Size**: `perimeter`, `equivalent_diameter`
- **Form**: `roundness`, `aspect_ratio`, `compactness`, `solidity`
- **Intensity / texture**: `mean_intensity`, `std_intensity`,
  `intensity_entropy`, `glcm_energy`, `glcm_correlation`, `lbp_mean`
- **Invariant moments**: `hu1`, `hu2`, `hu4`, `hu7`
- **Topology / morphology**: `euler_number`, `holes`, `convexity_defects`

Note that `roundness` and `compactness` are mathematically inverse
(`compactness = 4π / roundness`), and `euler_number` and `holes` are
perfectly redundant (`holes = 1 - euler_number`). These redundancies
are known and will be revisited after evaluating clustering results.
